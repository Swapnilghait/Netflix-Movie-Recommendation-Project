**IMPORTING LIBRARIES AND DATASET**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv("/content/drive/MyDrive/PREPINSTA /MACHINE LEARNING /6.ASSOCIATION RULE BASED LEARNING /DATASET /Netflix Movie Recommendation Ec.csv" , header = None)

# for storing all this transcations we need the list

transactions = [] # created the empty list

# for adding in transaction we will use the for loop here
for i in range (1 , 7466):
  transactions.append([str(data_set.values[i,j]) for j in range(0,20)])
```

```python
print(transactions)
```

**ECLET TRAINING ON DATASET**

```python
# importing the libraries
from apyori import apriori

# we will initialize it
rules = apriori (transactions = transactions , min_support = 0.003 , min_confidence = 0.0003 , min_lift = 3 , min_length = 2 , max_length = 2 )
```

**VISUALIZING ALL THE RESULT**

***RAW RESULT***

```python
# we need to create the list of basket intteligence

result = list(rules)
result
```

***PROPER DISPLAY***

```python
def inspect(result):
  Movie_1 = [tuple(result[2][0][0])[0] for result in result]
  Movie_2 = [tuple(result[2][0][1])[0] for result in result]
  support = [result[1] for result in result]
  return list(zip( Movie_1, Movie_2 , support))
resultsinDataFrame = pd.DataFrame(inspect (result) , columns = ["Movie 1" , "Movie 2" , "support"])
```
